# World Happiness Report 2017 – EDA

This notebook performs descriptive statistics and all common plots for the dataset.

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/mnt/data/happiness_2017.csv')
df.head()
```

## Dataset Info & Descriptive Statistics

```python

df.info()
```

```python

df.describe()
```

## Missing Values

```python

df.isnull().sum()
```

## Univariate Plots

```python

df.hist(figsize=(12,8))
plt.show()
```

## Boxplots

```python

df.plot(kind='box', figsize=(12,8))
plt.show()
```

## Correlation Heatmap

```python

corr = df.corr()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.show()
```

## Scatter Plot Example

```python

plt.scatter(df['Economy..GDP.per.Capita.'], df['Happiness.Score'])
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

## Bar Plot – Top 10 Happiest Countries

```python

top10 = df.sort_values('Happiness.Score', ascending=False).head(10)
plt.barh(top10['Country'], top10['Happiness.Score'])
plt.gca().invert_yaxis()
plt.show()
```
